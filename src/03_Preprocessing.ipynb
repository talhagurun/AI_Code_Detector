````python
import pandas as pd 
import re 

df = pd.read_csv('02_tekrar_eden_veriler_silindi_final_dataset_merged.csv')
print(f"Veri yüklendi : {len(df)} satır")

def preprocess_code(code):

    if not isinstance(code , str):
        return ""

    code = code.replace("```python" , "").replace("```", "")

    code = re.sub(r'(?m)^ *#.*\n?' , '' , code)
    code = re.sub(r'#.*' , '' , code)

    code = re.sub(r'""".*?"""', '', code, flags=re.DOTALL)
    code = re.sub(r"'''.*?'''", '', code, flags=re.DOTALL)

    code = re.sub(r'\n\s*\n', '\n', code)

    return code.strip()

print("Temizlik işlemi yapılıyor")
df['clean_code'] = df['code'].apply(preprocess_code)

initial_len = len(df)
df = df[df['clean_code'].str.len() > 10]
print(f"Sadece yorum olan {initial_len - len(df)} adet veri silindi")

print("TEMİZLİK KONTROLÜ")

sample_idx = df[df['label'] == 0].index[0]
print("Orijinal : \n")
print(df['code'].loc[sample_idx][:200])
print("Temizlenmiş : \n")
print(df['clean_code'].loc[sample_idx][:200])


df.to_csv("03_yorum_satirlari_silindi_final_dataset_merged.csv" , index = False)
````
